```python
import pandas as pd
import matplotlib.pyplot as plt

# Read CSV files into pandas DataFrames
df_macro = pd.read_csv("eval_data_macro.csv")
df_micro = pd.read_csv("eval_data_micro.csv")

# Filter and group data for MUTAG dataset and Prediction Accuracy metric
mutag_macro = (
    df_macro[
        (df_macro["Dataset"] == "MUTAG") & (df_macro["Metric"] == "Prediction Accuracy")
    ]
    .groupby("Model")["F1 Score"]
    .mean()
    .sort_index()
)
mutag_micro = (
    df_micro[
        (df_micro["Dataset"] == "MUTAG") & (df_micro["Metric"] == "Prediction Accuracy")
    ]
    .groupby("Model")["F1 Score"]
    .mean()
    .sort_index()
)

# Ensure that both series have the same models
common_models = mutag_macro.index.intersection(mutag_micro.index)

# Filter data for common models
mutag_macro_common = mutag_macro[common_models]
mutag_micro_common = mutag_micro[common_models]

# Plotting as line charts
fig, ax = plt.subplots()
ax.plot(
    mutag_macro_common.index,
    mutag_macro_common,
    label="MUTAG_macro",
    marker="o",
    linestyle="-",
)
ax.plot(
    mutag_micro_common.index,
    mutag_micro_common,
    label="MUTAG_micro",
    marker="o",
    linestyle="--",
)

ax.set_xticks(mutag_macro_common.index)
ax.set_xticklabels(
    mutag_macro_common.index, rotation=45
)  # Rotate for better readability
ax.set_ylabel("F1 Score")
ax.set_xlabel("Model")
ax.legend()
plt.tight_layout()  # Adjust layout for better fit
plt.savefig("eval_macro_micro/line_plot_f1_scores_pred_macro_micro_mutag.png", transparent = True, dpi=300)

plt.show()
```
